```python
import pandas as pd
df = pd.read_csv ('carprices.csv')
df.head(10)
```

```python
dm = pd.get_dummies(df['Car Model'])
dm
```

```python
merging =  pd.concat([df,dm],axis='columns')
merging
```

```python
x = merging.drop(['Car Model','Sell Price($)','Mercedez Benz C class'] ,axis = 'columns')
x
            
```

```python
y= merging['Sell Price($)']
y
```

```python
from sklearn import linear_model as lm
LR = lm.LinearRegression()
LR.fit(x,y)
```

```python
LR.score(x,y)
```

**Price of mercedez benz that is 4 yr old with mileage 45000**
